## Задание 2. 
Построение филогенетического дерева 10 ортологов гена Fox2P - факторая транскрипции из крупного семейтсва Fox. FOXP2 встречается у многих позвоночных, где он играет важную роль в имитации звуков у птиц (например, пением) и эхолокации у летучих мышей. FOXP2 также необходим для правильного развития речи и языка у людей. Самое интересное, что он есть и у анамний, которые в принципе звуки редко целенаправленно издают ( амфибии не в счёт). Дерево описывает относительно привычное эволюционное представление.
