```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = np.random.uniform(low=0, high=100, size=1000)
plt.figure(figsize=(8, 5))
sns.boxplot(x=data)
plt.title("Ящик с усами для равномерного распределения")
plt.xlabel("Значения")
plt.grid(True)
plt.show()
```

```python
data = np.array([
    79, 80, 80, 80, 74, 80, 80, 79, 64, 78,
    73, 78, 74, 45, 81, 48, 80, 82, 82, 70
])
q1 = np.percentile(data, 25)
q2 = np.percentile(data, 50)  # медиана
q3 = np.percentile(data, 75)

# (b) Межквартильный размах
iqr = q3 - q1
minimum = np.min(data)
maximum = np.max(data)

lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
out = data[(data < lower_bound) | (data > upper_bound)]

print("Минимум:", minimum)
print("Q1 (25%):", q1)
print("Медиана (Q2):", q2)
print("Q3 (75%):", q3)
print("Максимум:", maximum)
print("IQR:", iqr)
print("Выбросы:", out)

plt.figure(figsize=(8, 5))
sns.boxplot(x=data)
plt.title("Ящик с усами для заданных данных")
plt.xlabel("Значения")
plt.grid(True)
plt.show()
```
