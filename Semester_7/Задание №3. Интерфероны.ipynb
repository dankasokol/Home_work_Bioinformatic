## Образец 4114B

Интерферон β во всех образцах (6138B, 4114B, 3821B) вызывает больше статистически значимых изменений белков по сравнению с интерфероном α, что свидетельствует о более масштабной молекулярной перестройке клетки.

Во всех анализах значения надёжности идентификации белков (π2) для α и β практически совпадают (различия менее 1%), оба интерферона дают стабильные результаты. (ED) всегда выше для β, отражая более выраженное разделение между обработанными и контрольными клетками. Интерферон α вызывает однородный ответ внутри обработанных клеток (MED больше), то есть все клетки реагируют сходно; ответ на β более вариативен между отдельными клетками. По количеству активированных белков и их соотношению к подавленным, интерферон α демонстрирует более мощный эффект
